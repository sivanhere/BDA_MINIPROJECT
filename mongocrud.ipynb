## Aggregation

```python
agre=[{'$match':{'Rating': {'$gt':3}}},{'$group':{ '_id' : 0, 'sum':{ '$sum':1}}}]

ratingGreaterThanThree = list(col.aggregate(agre));
print("sum of product having rating greater than 3 are {}".format(ratingGreaterThanThree[0]['sum']))
```

```python
onlyClassAndDept = col.find({},{'_id':0,'Class Name':1,'Division Name':1})

for doc in onlyClassAndDept:
    print(doc)
```

```python
countType=[]
classType=['Dresses','Knits','Blouses','Sweaters','Pants']
for type in classType:
    agre=[{'$match':{'Class Name': type}},{'$group':{ '_id' : 0, 'sum':{ '$sum':1}}}]
    countType.append(list(col.aggregate(agre)));
    
for i,count in enumerate(countType):
    print("{} : {}".format(classType[i],count[0]['sum']))
```

```python
countType=[]
classType=['Dresses','Knits','Blouses','Sweaters','Pants']
for type in classType:
    agre=[{'$match':{'Class Name': type}},{'$group':{ '_id' : 0,'sum':{'$sum':'$Rating'}, 'entries':{ '$sum':1}}}]
    countType.append(list(col.aggregate(agre)));
    
for i,count in enumerate(countType):
    print("{} : {}".format(classType[i],(count[0]['sum']/count[0]['entries'])))
```

```python
countType=[]
classType=['Dresses','Knits','Blouses','Sweaters','Pants']
for type in classType:
    agre=[{'$match':{'$and':[{'Class Name': type} ,{'Recommended IND': 1} ]}},{'$group':{ '_id' : 0,'sum':{'$sum':1}}}]
    countType.append(list(col.aggregate(agre)));
    
print("recommended")
for i,count in enumerate(countType):
    print("{} : {}".format(classType[i],count[0]['sum']))
```

```python
countType=[]
classType=['Dresses','Knits','Blouses','Sweaters','Pants']
for type in classType:
    agre=[{'$match':{'$and':[{'Class Name': type} ,{'Recommended IND': 0} ]}},{'$group':{ '_id' : 0,'sum':{'$sum':1}}}]
    countType.append(list(col.aggregate(agre)));
    
print("not recommended")
for i,count in enumerate(countType):
    print("{} : {}".format(classType[i],count[0]['sum']))
```
